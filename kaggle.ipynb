```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set Seaborn style
sns.set_style("whitegrid")

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# -------------------
# Line Plot
# -------------------
plt.figure(figsize=(6, 4))
sns.lineplot(x=x, y=y, marker='o')
plt.title("Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()

# -------------------
# Pie Chart
# -------------------
labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 25]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.show()

# -------------------
# Scatter Plot
# -------------------
plt.figure(figsize=(6, 4))
sns.scatterplot(x=x, y=y, s=100, color='red')
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
